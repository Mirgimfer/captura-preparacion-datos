4. Creación de diccionario del padrón:




```python
# Creamos un diccionario a apartir del DataFrame de padrón
diccionario_padro ={}

# Iteramos sobre cada barrio único en el DataFrame
for barrio in df_2024_padro['Nom_Barri'].unique():
    # Filtramos los datos del barrio actual
    datos_barrio = df_2024_padro[df_2024_padro['Nom_Barri'] == barrio]

    # Sumamos los valores asegurando que sean numéricos y manejando NaN
    hombres = pd.to_numeric(datos_barrio[datos_barrio['SEXE'] == 1]['Valor'], errors='coerce').sum()
    mujeres = pd.to_numeric(datos_barrio[datos_barrio['SEXE'] == 2]['Valor'], errors='coerce').sum()

    # Aseguramos que los valores no sean NaN y convertirlos a enteros
    hombres = int(hombres) if not pd.isna(hombres) else 0
    mujeres = int(mujeres) if not pd.isna(mujeres) else 0
    total = hombres + mujeres

    # Guardamos los datos en el diccionario
    diccionario_padro[barrio] = {
        'Hombres': hombres,
        'Mujeres': mujeres,
        'Total': total
    }

import json

# Mostramos el diccionario utilizando "json.dumps" para visualizar de forma legible
print("Diccionario del padrón por barrios:")
print(json.dumps(diccionario_padro, indent=4, ensure_ascii=False))
```

5. Diccionario biblioteques:




```python
# Creamos un diccionario a apartir del DataFrame de biblioteques
biblioteques_dict = {}

# Iteramos sobre cada equipament
for equipament  in df_2024_biblioteques['Nom_Equipament'].unique():
  # filtramos los datos para el nombre equipamiento
  datos_equipament = df_2024_biblioteques[df_2024_biblioteques['Nom_Equipament']== equipament ]

  # Obtenenmos los valores para el equipamiento actual
  nom_districte = datos_equipament['Nom_Districte'].iloc[0]
  nom_barri = datos_equipament['Nom_Barri'].iloc[0]
  valor = datos_equipament['Valor'].sum()

  # Creamos el diccionario
  biblioteques_dict[equipament] = {
      'Nombre del distrito': nom_districte,
      'Nombre del barrio': nom_barri,
      'Vistas de la biblioteca a 2024': int(valor) if not pd.isna(valor) else 0

  }

# Mostrar el diccionario
print("Diccionario de las bibliotecas por equipamientos:")
print(json.dumps(biblioteques_dict, indent=4, ensure_ascii=False))
```

PARTE 3:




6. Combinando ambos diccionarios:




```python
# Creamos el diccionario combinado
diccionario_combinado = {}

for equipament, datos in biblioteques_dict.items():
    barrio = datos['Nombre del barrio']  # Extraemos el barrio de la biblioteca

    # Buscamos la información del padrón
    padro_info = diccionario_padro.get(barrio, "N/A")  # Si no existe, asignamos 'N/A'

    # Creamos la entrada en el nuevo diccionario
    diccionario_combinado[equipament] = {
        **datos,  # Mantiene los datos originales de la biblioteca
        'Padro': padro_info  # Agrega la información del padrón
    }

# Mostrar el diccionario combinado con formato legible
print("Diccionario combinado de bibliotecas y padrón:")
print(json.dumps(diccionario_combinado, indent=4, ensure_ascii=False))
```

7. Comprobación de registros:




Con la siguientes lineas de codigo vamos a comprar si el número de registros en el diccionario bibliotecas coincide con el número de instancias en el DataFrame original df_2024_bibliotoques.

```python
# Contamos el número de registros en el diccionario de bibliotecas
num_registros_diccionario = len(biblioteques_dict)

# Contamos el número de equipamientos únicos en el DataFrame original
num_instancias_df = df_2024_biblioteques['Nom_Equipament'].nunique()

# Comparamos ambos valores
print(f"Número de registros en el diccionario: {num_registros_diccionario}")
print(f"Número de equipamientos únicos en el DataFrame: {num_instancias_df}")
```

Como hemos podido comprobar, el numero es el mismo.

8. Obtencion de datos:




```python
# Obtenemos los datos de la 'Biblioteca Clara'
biblioteca_clara = diccionario_combinado.get('Biblioteca Clara', {})

# Extraemos la información del padrón
padro_info = biblioteca_clara.get('Padro', {})

# Extraemos la información de hombres, mujeres y visitas
hombres = padro_info.get('Hombres', 'N/A')
mujeres = padro_info.get('Mujeres', 'N/A')
visitas = biblioteca_clara.get('Vistas de la biblioteca a 2024', 'N/A')

# Mostrar la información
print(f"Biblioteca Clara:\nHombres: {hombres}\nMujeres: {mujeres}\nVisitas: {visitas}")
```

9. Calcular ratio:




```python
# Agregar el cálculo del Ratio
for biblioteca, datos in diccionario_combinado.items():
    padron = datos["Padro"]

    if padron != "N/A" and padron["Total"] > 0:
        datos["Ratio"] = datos["Vistas de la biblioteca a 2024"] / padron["Total"]
    else:
        datos["Ratio"] = "N/A"  # Si no hay datos de padrón, asignamos "N/A"

# Mostrar el diccionario con Ratio agregado
print(json.dumps(diccionario_combinado, indent=4, ensure_ascii=False))
```

10. Conversión de diccionario a DataFrame:




```python
# Convertimos el diccionario en DataFrame
df_ratio = pd.DataFrame([
    {"Biblioteca": biblioteca, "Ratio": datos["Ratio"]}
    for biblioteca, datos in diccionario_combinado.items()
])

# Mostramos el DataFrame
print(df_ratio)
```
